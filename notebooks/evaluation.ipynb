Self-contained downstream utility test for `golf_data.csv`.

```python
from pathlib import Path

import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

DATA_PATH = Path('golf_data.csv')
df = pd.read_csv(DATA_PATH)

TARGET_COL = 'Total'
FEATURE_COLS = [column for column in df.columns if column != TARGET_COL and pd.api.types.is_numeric_dtype(df[column])]
OFFICIAL_COLS = FEATURE_COLS + [TARGET_COL]

# Evaluate only on real rows that come from golf_data.csv.
df_pipeline_pool, df_real_test = train_test_split(df[OFFICIAL_COLS], test_size=0.15, random_state=42)

print(f'Loaded {len(df)} rows from {DATA_PATH.name}')
print(f'Training pool rows: {len(df_pipeline_pool)}')
print(f'Real test rows: {len(df_real_test)}')
```

```python
# ==========================================
# 4. THE DOWNSTREAM UTILITY TOURNAMENT (PURE TSTR)
# ==========================================
# Fallback values extracted from real data in case synthetic rows contain NaNs
feature_fill_values = df_pipeline_pool[FEATURE_COLS].apply(pd.to_numeric, errors='coerce').median(numeric_only=True)
target_fill_value = pd.to_numeric(df_pipeline_pool[TARGET_COL], errors='coerce').median()

# Clean and isolate the 15% real test set (remains completely pure)
X_test_real = df_real_test[FEATURE_COLS].apply(pd.to_numeric, errors='coerce')
y_test_real = pd.to_numeric(df_real_test[TARGET_COL], errors='coerce')
valid_test_mask = X_test_real.notna().all(axis=1) & y_test_real.notna()
X_test_real = X_test_real.loc[valid_test_mask]
y_test_real = y_test_real.loc[valid_test_mask]

if X_test_real.empty:
    raise ValueError('No valid real test rows remain after numeric cleaning. Check feature selection and dataset contents.')

# Define competitors. The baseline is real golf_data.csv training rows.
competitors = {
    '0. Real Dataset Baseline': df_pipeline_pool.copy(),
    '1. Pure Agentic Loop Data': df_synth_tabular_autodata,
}

tournament_results = []

print('\n' + '=' * 60)
print('TOURNAMENT START: EVALUATING SYNTHETIC QUALITY AGAINST UNSEEN REAL DATA')
print('=' * 60)

for name, df_synthetic_train in competitors.items():
    print(f'\nEvaluating: {name}')

    if df_synthetic_train is None or df_synthetic_train.empty:
        tournament_results.append({
            'Synthetic Training Method': name,
            'Total Training Rows': 0,
            'R2 Score (Higher is Better)': 'SKIPPED',
            'MAE (Lower Error)': 'SKIPPED',
            'RMSE': 'SKIPPED',
        })
        print('  skipped | no training rows provided')
        continue

    # Force alignment with the official golf_data.csv column scheme only.
    synthetic_aligned = df_synthetic_train.reindex(columns=OFFICIAL_COLS)

    X_train_synth = synthetic_aligned[FEATURE_COLS].apply(pd.to_numeric, errors='coerce')
    y_train_synth = pd.to_numeric(synthetic_aligned[TARGET_COL], errors='coerce')

    valid_train_mask = y_train_synth.notna()
    X_train_synth = X_train_synth.loc[valid_train_mask].fillna(feature_fill_values)
    y_train_synth = y_train_synth.loc[valid_train_mask].fillna(target_fill_value)

    if X_train_synth.empty:
        tournament_results.append({
            'Synthetic Training Method': name,
            'Total Training Rows': 0,
            'R2 Score (Higher is Better)': 'SKIPPED',
            'MAE (Lower Error)': 'SKIPPED',
            'RMSE': 'SKIPPED',
        })
        print('  skipped | no valid target values after cleaning')
        continue

    model = RandomForestRegressor(n_estimators=200, random_state=42)
    model.fit(X_train_synth, y_train_synth)

    predictions = model.predict(X_test_real)

    r2 = r2_score(y_test_real, predictions)
    mae = mean_absolute_error(y_test_real, predictions)
    rmse = np.sqrt(mean_squared_error(y_test_real, predictions))

    tournament_results.append({
        'Synthetic Training Method': name,
        'Total Training Rows': len(X_train_synth),
        'R2 Score (Higher is Better)': f'{r2:.4f}',
        'MAE (Lower Error)': f'{mae:.2f} yards',
        'RMSE': f'{rmse:.2f} yards',
    })
    print(f'  done | rows={len(X_train_synth)} | R2={r2:.4f} | MAE={mae:.2f} | RMSE={rmse:.2f}')

tournament_results_df = pd.DataFrame(tournament_results)
display(tournament_results_df)
```
